```python
#2.載入需要的library
import requests
from lxml import etree
from io import StringIO
import jieba
import pandas as pd
import time
import random
```

```python
#3.指定辭典檔
jieba.set_dictionary('./mydict/dict.txt.big')
#加入自定義辭典
jieba.load_userdict('./mydict/mydict.txt')
```

```python
#4.開檔
fileAllLines = []
#避免檔案過大造成記憶體不足使用 with open 及 read()控制讀取數量
with open('./file/healthy_all.txt','r',encoding="utf-8") as fileLine:
    for line in iter(lambda: fileLine.read(1024), ''):
        #print(line)
        fileAllLines.append(line)
new_line = ' '.join(fileAllLines)
#new_line
```

```python
#5.加入stop word
with open(file='./mydict/stop_words.txt', mode='r', encoding='utf-8') as file:
    stop_words = file.read().split('\n')
    
stop_words.append('\n')  ## 文章中有許多分行符號，這邊加入停用字中，可以把它拿掉
stop_words.append('\n\n')
#stop_words #查看停止字是否都有剔除
```

```python
#6.進行匯入停止字的分詞
start_time = time.time()

seg_stop_words_list = []
seg_words_list = jieba.lcut(new_line)
for term in seg_words_list:
    if term not in stop_words:
        seg_stop_words_list.append(term)

#7.查看分詞結果
#seg_stop_words_list 

print("--- spend %s seconds ---" % (time.time() - start_time))
```

```python
#8.用pandas建立dataframe
seg_df = pd.DataFrame(seg_stop_words_list, columns=['seg_word'])
seg_df
```

```python
#9.進行排序，查看詞頻最高的前20個詞
seg_df['count'] = 1
seg_freq_df = seg_df.groupby('seg_word').sum()
#seg_freq_df
print(seg_freq_df.sort_values(by=['count'],ascending=False).head(20))
```

```python
#10.將詞頻統計的dataframe存成csv檔
df_seg=seg_freq_df.sort_values(by=['count'],ascending=False)
df_seg.to_csv('frequencyNormal.csv',index=1)
```

# 文字雲步驟分解

1.載入需要的套件<br>

```python
#1.載入繪圖相關的library
from PIL import Image 
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
print(matplotlib.matplotlib_fname())
```

```python
#2.準備一個有經過stop word 處理的 counter
from collections import Counter
seg_stop_counter = Counter(seg_stop_words_list)
#seg_stop_counter #查看詞頻統計
```

## wordcloud 參數調整指南
參考此篇:https://www.itread01.com/content/1553252314.html

font_path=None:顯示的字體

width=400:顯示畫布的寬度，預設400像素

height=200:顯示畫布的高度，預設200像素

margin=2:畫布邊緣留白的空隙，預設留白空間是2像素

mask=None:是否使用圖片遮罩，預設不使用。若使用mask，則需提供一個二值化的mask（即只有0和1的黑白色mask），此時參數width和height會被忽略。

scale=1,在字段width和height乘以的倍數，最終呈現的畫布尺寸以這個結果。

background_color='white':設定背景顏色，預設為黑色。

max_words=200:單詞最多顯示數量，默認200個

```python
#3.用有處理stop words的斷詞跑文字雲

mask=np.array(Image.open(r'think1.png'))

wordcloud = WordCloud(
    scale=3,
    mask=mask,
    font_path='./font/TaipeiSansTCBeta-Regular.ttf',
    background_color='white',
    width=2400,
    height=2400
).generate_from_frequencies(seg_stop_counter)


plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
#plt.show() #查看文字雲
```
